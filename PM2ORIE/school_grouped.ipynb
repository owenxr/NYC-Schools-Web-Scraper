```python
import random

def find_best_values(model, data, columns, exp):
    best_probability = 0
    best_values = None
    max_expense = sum(exp)
    
    # Define the search space for each column
    search_space = { 
        'Classroom_Salary': (0, int(max_expense)//3),
        'Other_Inst_Salary': (0, int(max_expense)//3),
        'Inst_Benefits': (0, int(max_expense)//3),
        'Prof_Dev': (0, int(max_expense)//3),
        'Admin_Salary': (0, int(max_expense)//3),
        'Admin_Benefits': (0, int(max_expense)//3),
        'Other_Salaries': (0, int(max_expense)//3),
        'Other_Benefits': (0, int(max_expense)//3),
        'Other_Exp': (0, int(max_expense)//3)
    }
    
    # Perform random search for 100000 iterations
    for i in range(50000):
        # Generate a random set of values for each column
        values = { col: random.randint(*search_space[col]) for col in columns }
        
        # Check if total expense is less than or equal to max_expense
        if sum(values.values()) <= max_expense:
            # Prepare the input data for the model
            x = data.copy()
            for col, val in values.items():
                x[col] = val
            
            # Predict the probability of graduation
            prob = model.predict_proba(x)[0][1]
            
            # Check if this is the best set of values so far
            if prob > best_probability:
                best_probability = prob
                best_values = values
                
    return best_values, best_probability
```
