```python
import pandas as pd
import numpy as np
```

```python
graduation = pd.read_csv('graduation_data_full.csv')
```

```python
graduation = graduation.iloc[:, 1:]
graduation
```

```python
max_features = 6
n_estimators = 1000
```

```python
from statsmodels.miscmodels.ordinal_model import OrderedModel
from sklearn.ensemble import RandomForestClassifier

mean_graduation_rate = graduation['graduation_rate'].mean()

# perform a logistic regression on the filtered dataframe
X_rfc = graduation.drop(['School', 'graduation_rate', 'District'], axis=1)

print(X_rfc.columns)

y_rfc = graduation['graduation_rate'].apply(lambda x: 1.0 if x >= mean_graduation_rate else 0.0)
rfc = RandomForestClassifier(max_features=max_features, n_estimators=n_estimators)
rfc.fit(X_rfc, y_rfc)
score = rfc.score(X_rfc, y_rfc)
score
```

```python
from statsmodels.miscmodels.ordinal_model import OrderedModel
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# make predictions using random forest classification
y_pred_rfc = rfc.predict(X_rfc)

# calculate metrics for random forest classification
acc_rfc = accuracy_score(y_rfc, y_pred_rfc)
prec_rfc = precision_score(y_rfc, y_pred_rfc)
rec_rfc = recall_score(y_rfc, y_pred_rfc)
f1_rfc = f1_score(y_rfc, y_pred_rfc)

print('\nRandom Forest Classification Results:')
print(f'Accuracy: {acc_rfc}')
print(f'Precision: {prec_rfc}')
print(f'Recall: {rec_rfc}')
print(f'F1 Score: {f1_rfc}')
```

```python
cols = ['School_Total', 'male_pct', 'female_pct', 'asian_pct', 'black_pct', 'hispanic_pct', 'white_pct', 'multi_pct', 'native_pct', 'learning_english_pct', 'disability_pct', 'poverty_pct', 'homeless_pct', 'School_exp', 'Classroom_Salary', 'Other_Inst_Salary', 'Inst_Benefits', 'Prof_Dev', 'Admin_Salary', 'Admin_Benefits', 'Other_Salaries', 'Other_Benefits', 'Other_Exp', 'S_T_Ratio', 'lt_4', 'four_twenty', 'plus_21']
```

```python
new_data = graduation[graduation['School'] == "SOUTH BROOKLYN COMM HIGH SCHOOL"]
new_data = new_data.loc[:, cols]
print(new_data.columns)
```

```python
import random

def find_best_values(model, data, columns, exp):
    best_probability = 0
    best_values = None
    max_expense = sum(exp)
    
    # Define the search space for each column
    search_space = { 
        'Classroom_Salary': (0, int(max_expense)//3),
        'Other_Inst_Salary': (0, int(max_expense)//3),
        'Inst_Benefits': (0, int(max_expense)//3),
        'Prof_Dev': (0, int(max_expense)//3),
        'Admin_Salary': (0, int(max_expense)//3),
        'Admin_Benefits': (0, int(max_expense)//3),
        'Other_Salaries': (0, int(max_expense)//3),
        'Other_Benefits': (0, int(max_expense)//3),
        'Other_Exp': (0, int(max_expense)//3)
    }
    
    # Perform random search for 100000 iterations
    for i in range(10000):
        # Generate a random set of values for each column
        values = { col: random.randint(*search_space[col]) for col in columns }
        
        # Check if total expense is less than or equal to max_expense
        if sum(values.values()) <= max_expense:
            # Prepare the input data for the model
            x = data.copy()
            for col, val in values.items():
                x[col] = val
            
            # Predict the probability of graduation
            prob = model.predict_proba(x)[0][1]
            
            # Check if this is the best set of values so far
            if prob > best_probability:
                best_probability = prob
                best_values = values
                
    return best_values, best_probability
```

```python
y_pred_prob = rfc.predict_proba(new_data)[:, 1]

print(f"Probability of a graduation rate over the mean: {y_pred_prob[0]:.2f}")

best_values, best_probability = find_best_values(rfc, new_data, ['Classroom_Salary', 'Other_Inst_Salary', 'Inst_Benefits', 'Prof_Dev', 'Admin_Salary', 'Admin_Benefits', 'Other_Salaries', 'Other_Benefits', 'Other_Exp'], new_data['School_exp'])
```
